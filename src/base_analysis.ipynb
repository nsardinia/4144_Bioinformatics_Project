```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pathlib import Path
```

```python
DATA_PATH = Path("data\SRP120552.tsv").absolute()
```

```python
df = pd.read_csv(DATA_PATH, sep="\t")
```

```python
df
```

```python
df = df.set_index("Gene")
```

```python
log_scaled = np.log2(df + 1)
```

```python
log_scaled
```

```python
median_expression_range = log_scaled.median(axis=1, numeric_only=True)
```

```python
median_expression_range
```

```python
plot = median_expression_range.plot.density()
```

```python
print(f"Number of genes: {df.shape[0]}\nVariance: {median_expression_range.var()}")
```
